```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#Importing data
call_df = pd.read_csv("https://raw.githubusercontent.com/raviteja-padala/Six_Sigma/main/Bank_Call_Center_project/Six%20sigma%20-%20Bank%20call%20resolution%20dataset.csv")
call_df.set_index('SLNO')
```

## Process capability analysis is carried out to measure the ability of a process to meet the specifications.

* Process capability (Cp ): This is defined as the tolerance width divided by the total spread of
process (6 Sigma). <br>

* Cp indicates the spread of variation present in a process.

* Design Tolerance = USL - LSL

* Cp = Design_tolerance / (6 Ã— Standard deviation)



```python
#converting 'FCR%' , '5DR%' column from object data type to float data type
call_df['FCR%'] = call_df['FCR%'].str.rstrip('%').astype(float)
call_df['5DR%'] = call_df['5DR%'].str.rstrip('%').astype(float)
```

## Process capability of First call Resolution

```python
# Process capability of First call Resolution
FCR_USL  = 1     #100%
FCR_LSL =  0.75  #75%
Design_Tolerance = FCR_USL - FCR_LSL

FCR_mean = (call_df['FCR%'].mean())/100
FCR_standard_deviation = (call_df['FCR%'].std())/100

Cp = Design_Tolerance/(6*FCR_standard_deviation)
print(f'Cp value of First call Resolution  is {Cp}')

if Cp> 1:
  print(f'Cp value {Cp} of FCR is greater than 1, Hence process is capable as per Process capability analysis')
else:
  print(f'Cp value {Cp} of FCR is less than 1, Hence process is not capable as per Process capability analysis')
```

## Process capability of 5-Day  Resolution

```python
# Process capability of 5-Day  Resolution
USL  = 1     #100%
LSL =  0.90  #90%
Design_Tolerance_5DR = USL - LSL

mean_5DR = (call_df['5DR%'].mean())/100
standard_deviation_5DR = (call_df['5DR%'].std())/100

Cp_5DR  = Design_Tolerance_5DR/(6*standard_deviation_5DR)
print(f'Cp value of 5-day resolution is {Cp_5DR}')

if Cp_5DR> 1:
  print(f'Process capability value {Cp_5DR} of 5DR is greater than 1, Hence process is capable as per Process capability analysis')
else:
  print(f'process capabilty value {Cp_5DR} of 5DR is less than 1, Hence process is not capable as per Process capability analysis')
```

## Process Capability Index (Cpk): This is the capability index that accounts for the centering of the process and is defined as the minimum of Cpk upper and Cpk lower.

* Cpk is a measure of process performance capability.
Cpk indicates shifting or closeness of process average from the target or mean value

* Cpk = Minimum of Cpku and cpkl

* Cpku = USL - Mean / (3 x Standard deviation)

* Cpkl = Mean - LSL / (3 x Standard deviation)

## Process Capability Index (Cpk) of First call Resolution

```python
# Process capability Index of First call Resolution
FCR_USL  = 1     #100%
FCR_LSL =  0.75  #75%


FCR_mean = (call_df['FCR%'].mean())/100
FCR_standard_deviation = (call_df['FCR%'].std())/100


FCR_Cpku = FCR_USL - FCR_mean / (3*FCR_standard_deviation)
print(f"FCR Cpku value is {FCR_Cpku}")


FCR_Cpkl = FCR_mean- FCR_LSL / (3*FCR_standard_deviation)
print(f'FCR Cpkl value is {FCR_Cpkl}')


FCR_Cpk = min(FCR_Cpku, FCR_Cpkl)


print(f'Cpk value of First call Resolution  is {FCR_Cpk}')

if FCR_Cpk > 1:
  print(f'Cpk value {FCR_Cpk} of FCR is greater than 1, Hence process is capable as per Process capability analysis')
else:
  print(f'Cpk value {FCR_Cpk} of FCR is less than 1, Hence process is not capable as per Process capability analysis')
```

## Process Capability Index (Cpk) of 5-Day Resolution

```python
# Process capability Index of 5-Day  Resolution
USL_5DR  = 1     #100%
LSL_5DR =  0.90  #75%


mean_5DR = (call_df['5DR%'].mean())/100
standard_deviation_5DR = (call_df['5DR%'].std())/100


Cpku_5DR = USL_5DR - mean_5DR / (3*standard_deviation_5DR)
print(f"5DR Cpku value is {Cpku_5DR}")


Cpkl_5DR = mean_5DR - LSL_5DR / (3*standard_deviation_5DR)
print(f'5DR Cpkl value is {Cpkl_5DR}')


Cpk_5DR = min(Cpku_5DR, Cpkl_5DR)


print(f'Cpk value of 5-day resolution is {Cpk_5DR}')

if Cp_5DR> 1:
  print(f'Process capability index {Cpk_5DR} of 5DR is greater than 1, Hence process is capable as per Process capability analysis')
else:
  print(f'process capabilty index {Cpk_5DR} of 5DR is less than 1, Hence process is not capable as per Process capability analysis')
```

## Interpretations of Cp and Cpk:

* When the Cp value is greater than 1 i.e. Cp value >1: The process spread is less and all
products fall within the specification limit. Here the process is said to be quite capable of
meeting the specification limit.
* When the Cp value is equal to 1 i.e. Cp value =1: The process spread is little wide but
running within the designed specification limit. Here the process is said to be just capable
of meeting the specification limit.
* When the Cp value is less than 1 i.e. Cp value <1: The process spread is large and most of
the products fall outside the specification limit. Here the process is said to be incapable.
When Cp value equal to Cpk i.e. Cp=Cpk: The Process means is said to be at centre.
* When the Cpk value is less than 1 i.e. Cpk value <1: Indicates that the mean of the process
is shifted from target and defects will be produced.
* When the Cpk value is greater than 1 i.e. Cpk value >1: The center or mean of the process
may be shifted from target but still the process is capable of meeting design specification.
* In order to achieve Six Sigma quality in the organization, we must reduce the variation in
the process so as to achieve the desired value of Cp
